```python
"""
1. Student Marks Sorting (Bubble Sort)

Question: A teacher wants to arrange student marks in ascending order.

Input:

[75, 90, 65, 85, 70]

Output:

[65, 70, 75, 85, 90]
"""
```

```python
marks = list(map(int,input().split()))
n = len(marks)
for i in range(n):
  for j in range(i+1,n):
    if marks[i] > marks[j]:
      marks[i], marks[j] = marks[j], marks[i]
print(marks)
```

```python
"""
6. Card Sorting (Insertion Sort)

Question: Sort cards in a player's hand.

Input:

[7, 3, 9, 1, 5]

Output:

[1, 3, 5, 7, 9]
"""
```

```python
card_arr = list(map(int,input().split()))
n = len(card_arr)
for i in range(1,n):
  key = card_arr[i]
  j = i - 1
  while j >= 0 and card_arr[j] > key:
    card_arr[j+1] = card_arr[j]
    j -= 1
  card_arr[j+1] = key
print(card_arr)
```

```python
"""
9. Student Record Sorting (Merge Sort)

Question: Sort student records efficiently.

Input:

[55, 12, 89, 34, 23, 78]

Output:

[12, 23, 34, 55, 78, 89]
"""
```

```python
def merge_sort(arr):
  n = len(arr)
  if n <= 1:
    return arr

  mid = n // 2
  left = arr[:mid]
  right = arr[mid:]

  sorted_left = merge_sort(left)
  sorted_right = merge_sort(right)

  return merge(sorted_left,sorted_right)

def merge(left,right):
  result = []
  i, j = 0, 0

  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      result.append(left[i])
      i += 1
    else:
      result.append(right[j])
      j += 1

  result.extend(left[i:])
  result.extend(right[j:])

  return result

stud_arr = list(map(int,input().split()))
print(merge_sort(stud_arr))
```

```python
"""
11. Product Search Result Sorting (Quick Sort)

Question: Sort product prices.

Input:

[1000, 200, 1500, 500, 800]

Output:

[200, 500, 800, 1000, 1500]
"""
```

```python
def quick_sort(a):
  n = len(a)
  if n <= 1:
    return a

  left, right = [], []
  pivot = a[-1]

  for x in a[:-1]:
    if x <= pivot:
      left.append(x)
    else:
      right.append(x)

  return quick_sort(left) + [pivot] + quick_sort(right)

product_arr = list(map(int,input().split()))
print(quick_sort(product_arr))
```
