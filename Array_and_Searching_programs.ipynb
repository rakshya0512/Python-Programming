```python
"""
Question 1: Search Product ID

An online store maintains a list of product IDs.

Given a product ID, find whether it exists in the inventory.

If found, print its position; otherwise print "Product Not Found".

Input
5
101
205
309
412
515
309
Output
Found at Index 2
"""
```

```python
n = int(input())
product_id = list(map(int,input().split()))
target = int(input())
found = False
for i in range(len(product_id)):
  if product_id[i] == target:
    found = True
    print(f"Found at Index {i}")
    break
if not found:
  print("Product Not Found")
```

```python
"""
Scenario 1: ATM Card Verification
A bank stores valid ATM card numbers.
Given a card number, check whether it exists.
Input
5
1001
1002
1003
1004
1005
1003
Output
Card Valid
Hint
Array Sorted → Binary Search
"""
```

```python
n = int(input())
a = list(map(int,input().split()))
key = int(input())
low , high , flag = 0, len(a)-1, 0
while low <= high:
  mid = (low + high) // 2
  if a[mid] == key:
    flag = 1
    break
  elif a[mid] < key:
    low = mid + 1
  else:
    high = mid - 1
if flag == 1:
  print("Card Valid")
else:
  print("Card Invalid")
```
