```python
"""
1.Intelligent Numerical Processing Engine

A futuristic analytics organization continuously processes numerical datasets received from distributed computational systems.

During analysis, the processing engine must calculate the total sum of all elements present inside the array.

Design a Python program to find the sum of array elements.

Input Format
First line contains integer n
Second line contains n integers
Output Format

Print the sum of array elements.

Sample Input
5
1 2 3 4 5
Sample Output
15
"""
```

```python
# Sum of elements
n = int(input())
arr = list(map(int,input().split()))
sum = 0
for i in arr:
  sum += i
print(sum)
```

```python
"""
4.Secure Sequence Reversal Engine

A distributed cyber intelligence organization continuously processes encrypted numerical sequences.

During synchronization, the processing engine must reverse the order of array elements before transmission.

Design a Python program using arrays.

Sample Input
5
1 2 3 4 5
Sample Output
5 4 3 2 1
"""
```

```python
n = int(input())
arr = list(map(int,input().split()))
rev = []
for i in range(len(arr)-1,-1,-1):
  rev.append(arr[i])
for i in rev:
  print(i,end=" ")

#arr.reverse()
```

```python
"""
6.Autonomous Numerical Classification Engine

A next-generation computational laboratory processes numerical sequences for intelligent classification systems.

During analysis, the processing engine must count how many even and odd numbers are present inside the array.

Design a Python program using arrays.

Sample Input
6
1 2 3 4 5 6
Sample Output
Even: 3
Odd: 3
"""
```

```python
n = int(input())
arr =list(map(int,input().split()))
even_count,odd_count = 0,0
for i in arr:
  if i % 2 == 0:
    even_count += 1
  else:
    odd_count += 1
print("Even:",even_count)
print("Odd:",odd_count)
```

```python
"""
7.Distributed Data Replication System

A futuristic cloud storage organization continuously transfers numerical datasets between distributed storage units.

During replication, the synchronization engine must copy all elements from one array into another array.

Design a Python program using arrays.

Sample Input
5
1 2 3 4 5
Sample Output
1 2 3 4 5
"""
```

```python
"""
10.Intelligent Duplicate Elimination Engine

A distributed cyber analytics organization processes massive numerical datasets for secure synchronization systems.

During processing, the validation engine must remove duplicate elements from the array.

Design a Python program using arrays.

Sample Input
6
1 2 2 3 4 4
Sample Output
1 2 3 4
"""
```

```python
"""
Find pair of element from the given array that
when summed together will be equal
to the value of target variable.

INPUT
4
[ 7, 2, 11, 15]
9
OUTPTUT
7, 2
"""
```

```python
n = int(input())
arr = list(map(int, input().split()))
target = int(input())
found = False
for i in range(len(arr)):
  for j in range(len(arr)):
    if arr[i] + arr[j] == target and i != j:
      print(arr[i],arr[j])
      found = True
      break
  if found:
    break
  if not found:
    print("Pair Not Found")
```
