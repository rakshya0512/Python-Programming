```python
"""
1. Username Validation System
Scenario

A website allows usernames containing only letters and digits.

Task

Check whether a username is valid.

Input
user = "Rathna123"
Output
Valid Username
Hint

Use:

isalnum()
"""
```

```python
username = input()
if username.isalnum():
    print("Valid Username")
else:
    print("Invalid Username")
```

```python
"""
2. Password Strength Checker
Scenario

A company requires passwords to:

Have at least 8 characters
Contain one uppercase letter
Contain one digit
Task

Check if the password is strong.

Input
password = "Python123"
Output
Strong Password
"""
```

```python
password = input()
if password.isalnum() and len(password) >= 8 and any(c.isupper() for c in password):
  print("Strong Password")
else:
  print("Not a Strong Password")
```

```python
"""
3. Email Domain Extractor
Scenario

An HR system stores employee emails.

Task

Extract the domain name.

Input
email = "rathna@gmail.com"
Output
gmail.com
Hint

Use:

split("@")
"""
```

```python
email = input().split('@')
print(email[1])
```

```python
"""
5. Chat Message Cleaner
Scenario

A chat application removes extra spaces.

Input
msg = "   Hello    World   "
Output
Hello World
Hint
split()
join()
"""
```

```python
msg = input()
msg = msg.split()
msg = " ".join(msg)
print(msg)
```

```python
"""
6. Product Code Verification
Scenario

A warehouse stores product codes.

Task

Check whether the code starts with "PRD".

Input
code = "PRD5678"
Output
Valid Product
Hint
startswith()
"""
```

```python
"""
8. Word Frequency Counter
Scenario

Analyze customer reviews.

Input
review = "good service good quality good support"
Task

Count frequency of each word.

Output
good : 3
service : 1
quality : 1
support : 1
"""
```

```python
review = input().split()
temp = []
for i in range(len(review)):
  if review[i] not in temp:
    temp.append(review[i])
    print(f"{review[i]} : {review.count(review[i])}")
```

```python
"""
9. Secret Message Encryption
Scenario

Shift each character by one position.

Input
abc
Output
bcd
Example
a -> b
b -> c
c -> d
"""
```

```python
message = input()
for i in message:
  print(chr(ord(i)+1),end="")
```

```python
"""
10. Anagram Checker (Most Asked)
Scenario

Two encrypted messages are considered identical if they contain the same characters in different orders.

Input
s1 = "listen"
s2 = "silent"
Output
Anagram
"""
```

```python
s1 = input().lower()
s2 = input().lower()
if sorted(s1) == sorted(s2):
  print("Anagram")
else:
  print("Not Anagram")
```
