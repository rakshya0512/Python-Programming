```python
"""
Question 1

Create a class Person with:

name
age
display_person()

Create a class Student that inherits Person and adds:

roll_no
display_student()

Print all details.
"""
```

```python
class Person:
  def display_person(self):
    print(f"Person Details:\nName - {self.name}\nAge - {self.age}")
class Student(Person):
  def display_student(self):
    self.display_person()
    print(f"Roll No - {self.roll_no}")

s1 = Student()
s1.name = "Rakshya"
s1.age = 20
s1.roll_no = 230933
s1.display_student()
```

```python
"""
Question 3

Create the hierarchy:

Animal
   ↓
Dog
   ↓
Puppy

Methods:

Animal → eat()
Dog → bark()
Puppy → weep()

Create a Puppy object and call all methods.

"""
```

```python
class Animal:
    def eat(self):
        print("Animal eats....")
class Dog(Animal):
    def bark(self):
        print("Dog barks...")
class Puppy(Dog):
    def weep(self):
        print("Puppy Weeps...")

p1 = Puppy()
p1.eat()
p1.bark()
p1.weep()
```

```python
"""
Question 7

Create:

        Shape
      /   |   \
 Circle Rectangle Triangle

Each child class should have its own draw() method.

"""
```

```python
class Shape:
    def draw(self):
        print("Shapes are being drawn")

class Circle(Shape):
    def draw(self):
        print("⭕ Circle is being drawn")

class Rectangle(Shape):
    def draw(self):
        print("🟥 Rectangle is being drawn")

class Triangle(Shape):
    def draw(self):
        print("🔺Triangle is being drawn")

t = Triangle()
t.draw()

r = Rectangle()
r.draw()

c = Circle()
c.draw()
```

```python
"""
Question 9

Create:

        A
      /   \
     B     C
      \   /
        D

Methods:

A → methodA()
B → methodB()
C → methodC()
D → methodD()

Call all methods using D object.
"""
```

```python
"""
Question 10

Create a company structure:

          Employee
          /      \
      Manager   Developer
          \      /
          TeamLead

TeamLead should access methods from all parent classes.

"""
```

```python
class Employee:
    def emp(self):
        print("Employee is accessed")

class Manager(Employee):
    def manage(self):
        print("Manager is accessed")

class Developer(Employee):
    def dev(self):
        print("Developer is accessed")
class TeamLead(Manager,Developer):
    def lead(self):
        print("TeamLead is accessing")

t = TeamLead()
t.emp()
t.manage()
t.dev()
t.lead()
```

```python
"""
Question 11

Create:

Person
   ↓
Student

Use constructors in both classes and initialize data using super().

"""
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display_person(self):
        print(f"Person Details:\nName - {self.name}\nAge - {self.age}")

class Student(Person):
    def __init__(self,name,age,roll_no):
        super().__init__(name,age)
        self.roll_no = roll_no

    def display_student(self):
        self.display_person()
        print(f"Roll No - {self.roll_no}")

s1 = Student('Rakshya',20,230933)
s1.display_student()
```

```python
"""
Question 12

Create:

Vehicle
   ↓
Car

Use constructor inheritance and display:

brand
model
price
"""
```

```python
class Vehicle:
    def __init__(self,brand,price):
        self.brand = brand
        self.price = price

    def show_brand(self):
        print(f"Brand - {self.brand}")

    def show_price(self):
        print(f"Price - {self.price}")

class Car(Vehicle):
    def __init__(self,brand,model,price):
        super().__init__(brand,price)
        self.model = model

    def show_model(self):
        print(f"Car Details:")
        self.show_brand()
        print(f"Model - {self.model}")
        self.show_price()

c1 = Car("Maruti Suzuki","Ertiga",800000)
c1.show_model()
```

```python
"""
Question 13: Banking System
Account
   ↓
SavingsAccount

Parent:

account_number
balance

Child:

interest_rate

Display account information.
"""
```

```python
class Account:
    def __init__(self,acc_no,bal):
        self.acc_no = acc_no
        self.bal = bal

    def showBal(self):
        print(f"Account Details:\nAccount Number - {self.acc_no}\nBalance - {self.bal}")

class SavingsAccount(Account):
    def __init__(self,acc_no,interest,bal):
        super().__init__(acc_no,bal)
        self.interest = interest

    def showAcc(self):
        self.showBal()
        print(f"Interest rate (in %) - {self.interest} %")

s1 = SavingsAccount('76864516109273',3,23000)
s1.showAcc()
```

```python
"""
Question 14: School Management
Person
   ↓
Teacher

Parent:

name
age

Child:

subject

Display teacher details.
"""
```

```python
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def showName(self):
        print(f"Teacher Details:\nName - {self.name}\nAge - {self.age}")

class Teacher(Person):
    def __init__(self,name,age,sub):
        super().__init__(name,age)
        self.sub = sub

    def showDetails(self):
        self.showName()
        print(f"Subject Handled - {self.sub}")

t = Teacher('Kalaivani',38,'Chemistry')
t.showDetails()
```

```python
"""
Question 15: Online Shopping
Product
   ↓
Electronics

Parent:

product_name
price

Child:

warranty

Display complete product information.
"""
```

```python
class Product:
    def __init__(self,product_name,price):
        self.product_name = product_name
        self.price = price

    def showProduct(self):
        print(f"Product Details:\nProduct Name - {self.product_name}\nPrice - {self.price}")

class Electronics(Product):
    def __init__(self,product_name,price,warranty):
        super().__init__(product_name,price)
        self.warranty = warranty

    def showWarranty(self):
        self.showProduct()
        print(f"Warranty Period (in yrs) - {self.warranty} years")

t = Electronics('Electronic Kettle',1500,1)
t.showWarranty()
```
